```python
import pandas as pd
from pypinyin import lazy_pinyin
import re
import os
import pickle
```

```python
test_text = 'ab,b.b/b;b\'ab`b[b]b<b>中?b:b"b{b}b~ab!中华@b#b$b%b^b&b(b)b-b=b_b+b，b。b、b；ab‘b’b【b】b·b！b b…b（b）ab1010bAABBCb'
def split_text(text):
    pattern = r',|\.|/|;|\'|`|\[|\]|<|>|\?|:|"|\{|\}|\~|!|@|#|\$|%|\^|&|\(|\)|-|=|\_|\+|，|。|、|；|‘|’|【|】|·|！|:|：| |…|（|）|[0-9]+|[a-zA-Z]+|《|》|“|”|？|·|~|——|-'
    return [sentence for sentence in re.split(pattern, text) if len(sentence) > 1]
split_text(test_text)
```

```python
table_dir = './corpus/table/'
```

```python
def hanzi2id(table_dir):
    with open(table_dir+"hanzi2pinyin.csv") as fin:
        lines = fin.readlines()
    with open(table_dir+"hanzi2id.txt", "w+") as fout:
        for idx, line in enumerate(lines):
            line = line.strip().split(",")
            print(line[0], idx, file=fout)
hanzi2id(table_dir)
```

```python
def pinyin2id(table_dir):
    with open(table_dir+"pinyin2hanzi.csv") as fin:
        lines = fin.readlines()
    with open(table_dir+"pinyin2id.txt", "w+") as fout:
        for idx, line in enumerate(lines):
            line = line.strip().split(",")
            print(line[0], idx, file=fout)
pinyin2id(table_dir)
```

```python
hanzi2id_dict = {}
with open(table_dir+"hanzi2id.txt") as fin:
    lines = fin.readlines()
for idx, line in enumerate(lines):
    line = line.strip().split(" ")
    hanzi2id_dict[line[0]] = idx
hanzi2id_dict
```

```python
pinyin2id_dict = {}
with open(table_dir+"pinyin2id.txt") as fin:
    lines = fin.readlines()
for idx, line in enumerate(lines):
    line = line.strip().split(" ")
    pinyin2id_dict[line[0]] = idx
id2pinyin = list(pinyin2id_dict.keys())
```

```python
id2pinyin
```

```python
def pinyin2hanzi_id(table_dir):
    pinyin2hanzi = {}
    with open(os.path.join(table_dir, "pinyin2hanzi.csv")) as fin:
        lines = fin.readlines()
    for line in lines:
        line = line.strip().split(",")
        pinyin2hanzi[pinyin2id_dict[line[0]]] = [
            hanzi2id_dict[x] for x in line[1].split(' ')
        ]
    with open(table_dir+'pinyin2hanzi_id.pkl', 'wb') as handle:
        pickle.dump(pinyin2hanzi, handle, protocol=pickle.HIGHEST_PROTOCOL)
pinyin2hanzi_id(table_dir)
with open(table_dir+'pinyin2hanzi_id.pkl', 'rb') as handle:
    pinyin2hanzi_id_dict = pickle.load(handle)
print(pinyin2hanzi_id_dict)
```

```python
def hanzi2pinyin_id(table_dir):
    hanzi2pinyin = []
    with open(os.path.join(table_dir, "hanzi2pinyin.csv")) as fin:
        lines = fin.readlines()
    for line in lines:
        line = line.strip().split(",")
        hanzi2pinyin.append(pinyin2id_dict[line[1]])
    np.save(os.path.join(table_dir, "hanzi2pinyin_id"), np.array(hanzi2pinyin))


hanzi2pinyin_id(table_dir)
hanzi2pinyin = np.load(table_dir+"hanzi2pinyin_id.npy")
```

```python
def get_hanzi_id(hanzi, pinyin):
    i = 0
    while True:
        try:
            hanzi_id = hanzi2id_dict[hanzi+str(i)]
            if id2pinyin[hanzi2pinyin[hanzi_id]] == pinyin:
                return hanzi_id
        except:
            print(hanzi, pinyin)
            raise KeyError
        i+=1
```

```python
train_data = []
for line in data:
    ids = []
    sentence = line[0]
    pinyin = line[1]
    for hanzi, pinyin in zip(sentence, pinyin):
        if hanzi+"0" not in hanzi2id_dict:
            continue
        ids.append(get_hanzi_id(hanzi, pinyin))
    if len(ids) > 2:
        train_data.append(ids)
train_data
```

```python
with open(news_dir+"train_data.pkl", "wb") as handle:
    pickle.dump(train_data, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
